# 图1B

```python
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle        #用于打乱表格，使用于split函数之前
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import pandas as pd


def estimator_fig(X,Y,cv,random_state):
    """
    
    
    """
    # Create the RFE object and compute a cross-validated score.
    forest = RandomForestClassifier(random_state=random_state,oob_score=True,n_jobs=-1) 
    # The "accuracy" scoring is proportional to the number of correct
    # classifications

    min_features_to_select = 1  # Minimum number of features to consider
    rfecv = RFECV(estimator=forest, step=1, cv=StratifiedKFold(cv),           # 分层交叉验证
                  scoring='roc_auc',n_jobs=-1,
                  min_features_to_select=min_features_to_select)
    rfecv.fit(X, Y)

    print("Optimal number of features : %d" % rfecv.n_features_)

    # Plot number of features VS. cross-validation scores
    plt.figure(dpi=300)
    plt.xlabel("Number of features selected",fontsize=18)
    plt.ylabel("Cross validation score ",fontsize=18)
    plt.plot(range(min_features_to_select,len(rfecv.grid_scores_) + min_features_to_select),
             rfecv.grid_scores_,color="#2a557f",lw = 3)
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    bwith=3
    ax.spines['bottom'].set_linewidth(bwith)
    ax.spines['left'].set_linewidth(bwith)
    ax.tick_params(axis='both',labelsize=18)       # 调整处
    plt.savefig("Paper_Fig\\第二波作图\\图1B_2.svg",bbox_inches="tight") # svg图片的bug无法覆盖
    plt.show()
    score = rfecv.grid_scores_
    feature_choose = X.columns[rfecv.get_support() == True]     #这个会返回支持的True和False名单 rfecv.support_
    return score,   feature_choose   ,rfecv.ranking_         # 再返回一下选择得到的特征把  #为什么要返回score，是为了方便我们查看对应特征时候的数据，并且方便我们自己画图
    
```

```python
df_merge = pd.read_csv("Fig_data\\KS-T筛选后的数据.csv")

score,   feature_choose   ,ranking_ = estimator_fig(df_merge.iloc[:,2:],df_merge["Taste"],5,654)
```

# 图1E

```python
df_smote = pd.read_csv("SMOTE增强数据.csv")
```

```python
df_smote["Taste"]=0
for i in range(df_smote.shape[0]):
    if df_smote["taste_num"].iloc[i] == 1:
        df_smote["Taste"].iloc[i] = "Umami"
    else:
        df_smote["Taste"].iloc[i] = "Bitter"
df_smote["Label"]="Enhancing"
df_smote.drop("taste_num",axis=1,inplace = True)
df_smote
```

```python
from sklearn.manifold import Isomap, MDS, TSNE
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import StandardScaler
fig_names = ["PCA", "KPCA", "ISOMAP", "T-SNE", "MDS示"]
location = "Paper_Fig\\EDA\\"
```

```python
# 思路：先加载两个df，然后新增加一列 label标注他们的属性成分

df_original_tmp = pd.read_csv("Fig_data\\筛选过后的全部数据.csv")
df_original_x = df_original_tmp.iloc[:,2:]
df_original_y = df_original_tmp[["Taste"]]
df_original = df_original_x
df_original["Label"]="Original"
df_original["Taste"]=0
for i in range(df_original_y.shape[0]):
    if df_original_y["Taste"].iloc[i] == "Umami":
        df_original["Taste"].iloc[i] = "Umami"
    elif df_original_y["Taste"].iloc[i] == "Bitter":
        df_original["Taste"].iloc[i] = "Bitter"

# 先把两个df拼接在一起
df_duplicated = pd.concat([df_original,df_smote])  #.reset_index(level=0)
df_duplicated.drop_duplicates(df_original.columns[:-2],inplace=True,ignore_index = True)
df_duplicated
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def data_enhance_display_unsupport(Xtest,df_D,fig_names,location):
    """
    Xtest：去重的多维度特征数据
    df_D：与Xtest对应的 标签数据，两列，一个是Label 一个是最后一列表明是不是增强采样得到的
    fig_name: 绘制图像的名字
    """
#     实例化降维器
    pca = PCA(n_components = 2,random_state = 123)
    kpca = KernelPCA(n_components = 2,kernel = "poly", ## 核函数为rbf核
                 gamma = 0.2,random_state = 123)
    ## 流行学习进行数据的非线性降维
    isomap = Isomap(n_neighbors = 2,                    ## 每个点考虑的近邻数量
                    n_components = 2)                   ## 降维到3维空间中
    tsne = TSNE(n_components = 2,perplexity =25,        ## TSNE进行数据的降维,降维到3维空间中
                early_exaggeration =3,random_state=123) 
    mds = MDS(n_components = 2,dissimilarity = "euclidean",random_state=123)   ## MDS进行数据的降维,降维到3维空间中

    model_estimators = [pca, kpca,isomap,tsne,mds]
#     model_estimators = [isomap,tsne,mds]
    
    Xtest_tmp = StandardScaler().fit_transform(Xtest)
    for (name, method) in zip(fig_names, model_estimators): #五种循环的方法
#         print(Xtest)
        if name == "PCA":
            
            tmp_X = method.fit_transform(Xtest_tmp)
            X_test_MDS = pd.DataFrame(tmp_X,columns=["Features_1","Features_2"])

            df_fig = pd.concat([df_D,X_test_MDS],axis=1)
            plt.figure(dpi= 300)

            #尝试修改一下颜色
            pall = sns.color_palette(['#2a557f', '#f05076'])  # ['#2a557f','#45bc9c','#f05076','#ffcd6e']
            sns.set_palette(pall)

            sns.scatterplot(data=df_fig, x="Features_1", y="Features_2", hue="Taste", style="Label",s=55)
            ax = plt.gca()
            bwith=3
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.spines['bottom'].set_linewidth(bwith)
            ax.spines['left'].set_linewidth(bwith)


            ax.legend(loc="lower right")
            plt.xticks(fontsize=18); plt.yticks(fontsize=18)
    #         plt.title(name+" Display", fontsize=18)
            plt.xlabel("PCA1",fontsize=18)
            plt.ylabel("PCA2",fontsize=18)
            plt.savefig(location+name+".svg",bbox_inches="tight")
            plt.show()
        
        
fig_names = ["PCA","KPCA","ISOMAP","T-SNE","MDS"]   
location = "Paper_Fig\\第二波作图\\"
data_enhance_display_unsupport(df_duplicated.iloc[:,:-2],df_duplicated,fig_names,location)       
```
