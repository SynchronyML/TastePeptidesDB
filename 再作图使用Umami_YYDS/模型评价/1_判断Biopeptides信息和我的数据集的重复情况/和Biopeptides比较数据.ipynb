```python
dict_name = {}
Pepname_list = []
for i,name_tmp in zip(range(df_biopeptides.shape[0]),df_biopeptides["PepName"]):
    name="".join(name_tmp.split())  # 去掉末尾空格
    Taste = df_biopeptides["Taste"].iloc[i]
    Taste_list = Taste.split(" ")             # ['Bitter', 'peptide', '']
    # name 是 序列号，df_biopeptides["Taste"] 是 几个滋味
    if name not in Pepname_list:
        dict_name[name] = [Taste_list[0]]                 # 不存在直接新建
        Pepname_list.append(name)
            
    else:
        tmp = dict_name[name]
        tmp.append(Taste_list[0])
        dict_name[name] = tmp               # 存在的话，直接加入进入
    
    
dict_name        
```

# 开始比较

## 定义编码函数
允许他的值，或者滋味比我出现的少

```python
## 比较两个列表A和B是否A包含于B，可以比较有间隔的列表
def allin(A, B):
    """
    判断A在B里面
    a = ["Umami","Bitter"]
    b = ["Bitter","Sour","Umami"]
    True
    """
    A.sort()  # 因为排序了，所以对于先后顺序没有要求了
    B.sort()
    if len(A) > len(B): return False
    k = 0
    for i in range(0,len(A)):
        for j in range(k,len(B)):
            if A[i] == B[j]:
                if i == len(A)-1:
                    return True
                else:
                    break
        if j < len(B)-1:
            k = j+1
        else:
            break
    return False

def taste_compare(str_1,str_2): 
    """
    一定要是str_2 ，包含字符串str_1
    """
    str_2 = str_2.split(",")
    
    return allin(str_1,str_2)
```

```python
taste_add = []
taste_update=[]
for key,value in dict_name.items():
    try:
        taste_name = my_dict["Taste"][key]
        if taste_compare(value,taste_name): # 如果全部包含
            pass
        else:
            print(f"存在但滋味不相同{key}")
            taste_update.append(key)
    except:
            print(f"不存在{key}")
            taste_add.append(key)
```

```python
data_add = pd.DataFrame(np.array(taste_add),columns=["PepName"])
# 写一个循环，第二列是Biopeptides的滋味，第三列是我们已有的滋味属性
data_add["Biopeptides_Taste"] = "GG"
for i in range(data_add.shape[0]):
    data_add["Biopeptides_Taste"].iloc[i]=dict_name[data_add["PepName"].iloc[i]]

data_add.to_csv("待录入的鲜味肽.csv")
```
