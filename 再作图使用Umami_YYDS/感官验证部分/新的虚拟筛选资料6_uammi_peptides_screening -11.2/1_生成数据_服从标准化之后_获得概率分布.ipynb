# 生成数据

```python
letter = ["A","R","N","D","C","Q","E","G", "H","I","L","K","M","F","P","S","T","W","Y","V"]
```

## 生成2肽

```python
Two_peptides = []
for i in letter:
    for ii in letter:
        tmp = i+ii
        if tmp not in target:
            Two_peptides.append(tmp)
Two_peptides
```

## 生成3肽

```python
Three_peptides = []
for i in letter:
    for ii in letter:
        for iii in letter:
            tmp = i+ii+iii
            if tmp not in target:
                Three_peptides.append(tmp)
Three_peptides
```

## 合并并保存df_merge为Y

```python
Two_peptides.extend(Three_peptides)
df_merge = pd.DataFrame(Two_peptides,columns=["Seq"])
df_merge.to_csv("1_未经过标准化8177_Candidate_Label.csv",index = False)    #未经过标准化
df_merge
```

# 制作潜在搜索部分
## 生成9大特征

```python
from rdkit import Chem
import os
import pandas as pd
import numpy as np
from rdkit.ML.Descriptors import MoleculeDescriptors
from rdkit.Chem import Descriptors
from icecream import ic
from time import time 
def AA_to_des(dateframe_name):
    mols = []
    for i in dateframe_name["Seq"]:
        mol = Chem.MolFromFASTA(i)
        mols.append(mol)

    descs = [desc_name[0] for desc_name in Descriptors._descList]

    # 展示一下所有的特征名单
    print(f"共有{len(mols)}个化合物", end="______");
    print(f"共有{len(descs)}个特征")
    print(f"特征如下：{[i for i in descs]}")

    #计算描述符
    desc_calc = MoleculeDescriptors.MolecularDescriptorCalculator(descs)
    descriptors = pd.DataFrame(desc_calc.CalcDescriptors(mol) for mol in mols)
    descriptors.columns = descs

    #使用方差来选择特征,m目的是为去除全0特征
    from sklearn.feature_selection import VarianceThreshold
    X = descriptors.values
    select = VarianceThreshold()
    X_new = select.fit_transform(X)
    
    # get_support可以返回被选中特征和没被选中特征的掩码或者整数索引，这里是False就是返回被剔除
    print(np.array(descs)[select.get_support()==False])

    desc_new_df = pd.DataFrame(X_new, columns=np.array(descs)[select.get_support()==True])
    print(desc_new_df.head())
    print(desc_new_df.shape)
    
    return desc_new_df
a = time()
result = AA_to_des(df_merge)
b = time()
print(b-a)
result
```
