```python
import joblib
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.utils import shuffle 
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, auc,roc_auc_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# 读取数据并标准化

```python
estimator3 = joblib.load('split_GTB_2.pkl')
estimator3
train_X = pd.read_csv("train_X_2.csv")
train_Y = pd.read_csv("train_Y_2.csv")
test_X = pd.read_csv("test_X_2.csv")
test_Y = pd.read_csv("test_Y_2.csv")
# GtBossT = GradientBoostingClassifier(random_state=0,criterion='friedman_mse',#learning_rate = 1.1,
#                                    loss = 'deviance', max_features = 'sqrt',
#                                    min_samples_split = 2,n_estimators = 130)
GtBossT = estimator3
GtBossT.fit(train_X,train_Y)
gtb_train = GtBossT.predict(train_X)
gtb_test = GtBossT.predict(test_X)
print(f"测试集准确度是{accuracy_score(train_Y,gtb_train)}")
print(f"验证集准确度是{accuracy_score(test_Y,gtb_test)}")

train_matrix = confusion_matrix(train_Y,gtb_train)
test_matrix = confusion_matrix(test_Y,gtb_test)

sns.set_style("whitegrid")
#尝试修改一下颜色
pall = sns.color_palette(['#73a2c6', '#f4777f'])
sns.set_palette(pall)

# 绘制热图
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.heatmap(train_matrix,annot=True,cmap="Blues",cbar=None,fmt="d")  #cbar=None 就会不显示色条,fmt="d"确保中间现实数字
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.title("Confusion Matrix of Train")

plt.subplot(1,2,2)
sns.heatmap(test_matrix,annot=True,cmap="Blues",cbar=None,fmt="d")  #cbar=None 就会不显示色条,fmt="d"确保中间现实数字
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.title("Confusion Matrix of Test")
plt.show()

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
# 绘制训练集ROC图
y_pro_train = GtBossT.predict_proba(train_X)[:,1]
# 获取预测
y_pre_train = GtBossT.predict(train_X)
#计算真阳和假阳概率
fpr, tpr, thresholds = roc_curve(train_Y,y_pro_train)
roc_auc = auc(fpr, tpr)  #也等于roc_auc_score(y_test,y_pro)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr,label='AUC = %0.2f'% roc_auc)
plt.legend(loc="best")
plt.plot([0,1],[0,1],'--')
# plt.xlim([-0.1,1.1])
# plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')


# 绘制测试集ROC图
plt.subplot(1,2,2)
# 获取预测的概率
y_pro_test = GtBossT.predict_proba(test_X)[:,1]
# 获取预测
y_pre_test = GtBossT.predict(test_X)

#计算真阳和假阳概率
fpr, tpr, thresholds = roc_curve(test_Y,y_pro_test)
roc_auc = auc(fpr, tpr)  #也等于roc_auc_score(y_test,y_pro)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr,label='AUC = %0.2f'% roc_auc)
plt.legend(loc="best")
plt.style.use("seaborn")
plt.plot([0,1],[0,1],'--')
# plt.xlim([-0.1,1.1])
# plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
# plt.savefig("ROC.jpg", dpi = 300)
plt.show()
```

## 实例化一个GTBoost

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.utils import shuffle 
import seaborn as sns
sns.set_style("whitegrid")
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection


df = pd.read_csv("data_smote.csv")
df = shuffle(df)
X = df.iloc[:,1:]
Y = df["Label"]
columns = X.columns
X = StandardScaler().fit_transform(X)
X = pd.DataFrame(X,columns = columns)

GtBossT = GradientBoostingClassifier(random_state=0,criterion='friedman_mse',learning_rate = 2.4,
                                   loss = 'exponential', max_features = 'sqrt',
                                   min_samples_split = 4,n_estimators = 150)
GtBossT.fit(X,Y)
```

## pdp_isolate
Y轴表示预测相较于基准线或最左值的增加值
蓝色区域表示置信区间
从上图针对进球数的PDP分析看，随着进球数增多，球员获得“足球先生”的概率也会逐步增加，但增加到一定程度之后就收敛了

```python
df = pd.read_csv("data_smote.csv")
df = shuffle(df)
X = df.iloc[:,1:]
Y = df["Label"]
columns = X.columns

GtBossT = GradientBoostingClassifier(random_state=0,criterion='friedman_mse',learning_rate = 2.4,
                                   loss = 'exponential', max_features = 'sqrt',
                                   min_samples_split = 4,n_estimators = 150)
GtBossT.fit(X,Y)

feat_name =  'BCUT2D_LOGPHI'
nick_name = 'BCUT2D_LOGPHI'

pdp_dist = pdp.pdp_isolate(
    model = GtBossT, dataset = df, model_features = columns, feature=feat_name
)

fig, axes = pdp.pdp_plot(pdp_dist, nick_name, plot_lines=True, frac_to_plot=0.8, plot_pts_dist=True)
```

```python
# 划分训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)

y_pred = GtBossT.predict(X_test)
y_pred_proba = GtBossT.predict_proba(X_test)
```

```python
import shap
shap.initjs()

explainer = shap.TreeExplainer(GtBossT)
shap_values = explainer.shap_values(X_test)   

```

## 展示某个特征从小变大时对预测结果的shap值。
Partial Dependence Plot

纵轴：是预测结果
<br>
纵轴平均线：模型预测的平均结果
<br>
横轴：特征数据点大小
<br>
横轴虚线：数据点平均大小

```python
for i in columns:
    shap.partial_dependence_plot(i, GtBossT.predict, X_test, model_expected_value=True, feature_expected_value=True)
```
