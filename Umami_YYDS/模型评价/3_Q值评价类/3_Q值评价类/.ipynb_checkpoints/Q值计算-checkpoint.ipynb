```python
import pandas as pd
import numpy as np
```

```python
df_tmp = pd.read_csv("爬取结果进行比较.csv")
df_tmp["Q_Value"] = -1
df_tmp["Q"] = -1
df_tmp
```

```python
for i in range(df_tmp.shape[0]):
    Taste_list = df_tmp["PepName"].iloc[i]
    value = 0
    for aa in list(Taste_list):
        if aa == "A":
            value += 731.034
        elif aa == "C":
            value += 1000.991
        elif aa == "D":
            value += 539.914
        elif aa == "E":
            value += 549.47
        elif aa == "F":
            value += 2599.232
        elif aa == "H":
            value += 499.301    
        elif aa == "I":
            value += 2969.527
        elif aa == "K":
            value += 1500.292
        elif aa == "L":
            value += 2420.057
        elif aa == "M":
            value += 1299.616
        elif aa == "N":
            value += 86.004 
        elif aa == "P":
            value += 2649.401
        elif aa == "Q":
            value += -102.727
        elif aa == "R":
            value += 731.034  
        elif aa == "S":
            value += 40.613
        elif aa == "T":
            value += 439.576
        elif aa == "V":
            value += 1689.023
        elif aa == "W":
            value += 301.014
        elif aa == "Y":
            value += 2869.189
#     print(len(list(Taste_list)))
    tmp = value / len(list(Taste_list))
    df_tmp["Q_Value"].iloc[i] = tmp
    if tmp <= 1200:
        df_tmp["Q"].iloc[i] = 1
    elif tmp >= 1400:
        df_tmp["Q"].iloc[i] = 0

df_tmp.to_csv("加入Q值的判断.csv",index=False)
df_tmp

            
```
